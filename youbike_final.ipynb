```python
def search_for_nearest_new(df_geo_3857,user_lat_4326,user_lon_4326,user_loc_3857,mode,action,n=5):

    api_keys='AIzaSyB86TP6h4aMMtS8ywWUbpmWdEue3Y875bI'
    gmaps=googlemaps.Client(key=api_keys)
    
    k_neighbour= n # ball tree search for n nearest point

    # (epsg:3857)
    ### ball tree search # 離找最近的n個ubike站點

    balltree_candidates = [(point.x, point.y) for point in df_geo_3857.geometry]
    candidates = np.array(balltree_candidates)
    #display(candidates)
    ball_tree=BallTree(candidates, leaf_size=30, metric='euclidean')
    distances, indices = ball_tree.query(user_loc_3857, k=k_neighbour)
    indices = indices.flatten()
    close_5_df=df_geo_3857.iloc[indices]
    close_5_df=close_5_df.reset_index(drop=True)

    #display(close_5_df)

    #
    
    # epsg:4326 
    # 根據需求（還車return/借車rent）or 全套both *還車bicycling/借車walking*
    # 找最近 且 車太少/車太多的
    # 利用googleapi 計算出直綫距離中 最近的5個點 
    # 騎車/步行 真實路徑距離 找出最短的

    # mode = walking / bicycling (to calculate travel time)

    # action - True: empty percentage 最少（最空，要還車）False viceversa

    origin = [(user_lat_4326,user_lon_4326)]  # 用戶的經緯度

    destination=[] # 記錄目標點

    for index in range(len(close_5_df)):
        
        destination.append((close_5_df.loc[index,'latitude_4326'],close_5_df.loc[index,'longitude_4326'])) 
        # 將最靠近的5個點建立起list，作爲 google distance matrix input


    result= gmaps.distance_matrix(origin,destination,mode=mode)
    for i, row in enumerate(result['rows']): # 時間計算
        for j, element in enumerate(row['elements']):
            close_5_df.loc[j,'route_distance']=element['distance']['value']
            close_5_df.loc[j,'route_time']=element['distance']['value']
            distance = element['distance']['value']
            time = element['duration']['value']
            #print(f"    到点{j+1}的距离为: {distance}")
            
    shortest_5=close_5_df.sort_values(by='route_distance',ascending=True)
    shortest_5=shortest_5.reset_index(drop=True)
    shortest_5_withpoint=close_5_df.sort_values(by='empty_percentage',ascending=action)
    shortest_5_withpoint=shortest_5_withpoint.reset_index(drop=True)
    shortest_5=shortest_5.to_crs('epsg:4326')
    shortest_5_withpoint=shortest_5_withpoint.to_crs('epsg:4326')

    #print(shortest_5)
    #print()
    #print(shortest_5_withpoint) # 最接近的五個點中 最缺/多車
        
    return shortest_5,shortest_5_withpoint

```
